# Day 001: Vectors and Tensors


## 1. Objective
To represent real-world data as mathematical vectors and implement fundamental vector arithmetic (addition and scalar multiplication) from scratch to understand how data scales and combines in dimensional space.

## 2. Mathematical Foundation
Here is the formal notation for what we are about to build.  
A **vector** is a 1-dimensional array of numbers.  
A **tensor** is the generalized term for an $n$-dimensional array (a scalar is 0D, a vector is 1D, a matrix is 2D).  
***Vector Addition***: > Two vectors $A$ and $B$ must occupy the same dimensional space (have the same length $n$). We add them by summing their corresponding components.  

$$A + B = \begin{bmatrix} a_1 \\ a_2 \\ \vdots \\ a_n \end{bmatrix} + \begin{bmatrix} b_1 \\ b_2 \\ \vdots \\ b_n \end{bmatrix} = \begin{bmatrix} a_1 + b_1 \\ a_2 + b_2 \\ \vdots \\ a_n + b_n \end{bmatrix}$$

**Scalar Multiplication**: > Multiplying a vector $A$ by a scalar $c$ (a single number) scales the vector's magnitude without changing its fundamental direction.  
$$c \cdot A = c \begin{bmatrix} a_1 \\ a_2 \\ \vdots \\ a_n \end{bmatrix} = \begin{bmatrix} c \cdot a_1 \\ c \cdot a_2 \\ \vdots \\ c \cdot a_n \end{bmatrix}$$


## 3. Algorithmic Strategy
To build this in pure Python, we cannot just write A + B with standard Python lists (Python would just concatenate the lists, e.g., [1] + [2] = [1, 2], not [3]).

Addition Strategy: We will use Python's built-in **zip()** function. zip() pairs up elements from two lists, allowing us to iterate through them simultaneously—much like iterating through related records in a Django QuerySet. We will use a list comprehension to add the pairs together.

Scalar Strategy: We will iterate through a single list, multiplying each element by the scalar value, again using a clean list comprehension.

## 4. Implementation

```python
import numpy as np

# Write your raw Python/NumPy code here

class Vector:
    def __init__(self, elements):
        """Initialize with a list of numbers."""
        self.elements = elements 
    
    def __repr__(self):
        return f"Vector({self.elements})"
    
    def __add__(self, other):
        """Element-wise addition using zip() and list comprehensions."""
        if len(self.elements) != len(other.elements):
            raise ValueError("Mathematical Error: Vectors must be the same length.")
        
        #zip(A, B) creates pairs: (a1, b1), (a2, b2)
        summed_elements = [a+b for a, b in zip(self.elements, other.elements)]
        return Vector(summed_elements)
    
    def __mul__(self, scalar):
        "Element-wise scalar multiplication"
        scaled_elements = [a*scalar for a in self.elements]
        return Vector(scaled_elements)
```

## 5. Verification

```python
# --- 1. Testing our Custom Class ---
# Inventory: [Milk, Butter, Cheese, Yogurt]
ktm_vector = Vector([500, 150, 45, 80])
ilam_vector = Vector([300, 90, 60, 120])

print("--- Custom Pure Python Vectors ---")
total_inventory = ktm_vector + ilam_vector
print(f"Total Inventory (Addition): {total_inventory}")

projected_ktm = ktm_vector * 1.5
print(f"Projected KTM (+50% Growth): {projected_ktm}\n")

# --- 2. The Professional Standard: NumPy Tensors ---
# NumPy natively understands element-wise operations on n-dimensional arrays (tensors).
ktm_tensor = np.array([500, 150, 45, 80])
ilam_tensor = np.array([300, 90, 60, 120])

print("--- NumPy 1D Tensors ---")
np_total = ktm_tensor + ilam_tensor
np_projected_ktm = ktm_tensor * 1.5 

print(f"NumPy Total: {np_total}")
print(f"NumPy Ktm projected (+50% Growth): {np_projected_ktm}")


# Verifying our scratch code matches the C-optimized library
assert total_inventory.elements == np_total.tolist(), "Addition logic failed!"
assert projected_ktm.elements == np_projected_ktm.tolist(), "Scalar logic failed!"

assert np.allclose(total_inventory.elements, np_total.tolist(), atol=1e-8), "Addition Logic Failed!"
assert np.allclose(projected_ktm.elements, np_projected_ktm.tolist(), atol=1e-8), "Scalar Logic Failed!"

print("\n Verification Passed: Custom Logic matches Numpy Mechanics Perfectly. ")
```
