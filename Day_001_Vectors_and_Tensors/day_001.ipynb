## 2. Mathematical Foundation
Here is the formal notation for what we are about to build.  
A **vector** is a 1-dimensional array of numbers.  
A **tensor** is the generalized term for an $n$-dimensional array (a scalar is 0D, a vector is 1D, a matrix is 2D).  
***Vector Addition***: > Two vectors $A$ and $B$ must occupy the same dimensional space (have the same length $n$). We add them by summing their corresponding components.  

$$A + B = \begin{bmatrix} a_1 \\ a_2 \\ \vdots \\ a_n \end{bmatrix} + \begin{bmatrix} b_1 \\ b_2 \\ \vdots \\ b_n \end{bmatrix} = \begin{bmatrix} a_1 + b_1 \\ a_2 + b_2 \\ \vdots \\ a_n + b_n \end{bmatrix}$$

**Scalar Multiplication**: > Multiplying a vector $A$ by a scalar $c$ (a single number) scales the vector's magnitude without changing its fundamental direction.  
$$c \cdot A = c \begin{bmatrix} a_1 \\ a_2 \\ \vdots \\ a_n \end{bmatrix} = \begin{bmatrix} c \cdot a_1 \\ c \cdot a_2 \\ \vdots \\ c \cdot a_n \end{bmatrix}$$
