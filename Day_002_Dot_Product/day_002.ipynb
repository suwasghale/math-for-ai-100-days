# Day 002: The Dot Product (Scalar Product)


## 1. Objective
> The **Dot Product** is the engine of similarity and reduction in linear algebra. Unlike element-wise operations (Addition, Hadamard) that return a new vector, the dot product takes two vectors of equal length collapses them into a **single scalar number**, representing *how much those two vectors align or interact with each other*.

## 2. Mathematical Foundation
> The dot product computes the sum of the products of the corresponding entries of two sequences of numbers.

Let $A$ and $B$ be two vectors of length $n$.

**The Algebraic Definition (Computation):**
$$A \cdot B = \sum_{i=1}^{n} A_i B_i = (A_1 \times B_1) + (A_2 \times B_2) + \dots + (A_n \times B_n)$$

**The Geometric Definition (Intuition):**
It measures the directional agreement between two vectors in space, where $\theta$ is the angle between them:
$$A \cdot B = \|A\| \|B\| \cos(\theta)$$
* If $\theta = 0^\circ$, $\cos(0) = 1$ (Maximum alignment).
* If $\theta = 90^\circ$, $\cos(90) = 0$ (Orthogonal / no similarity).

## 3. Real-World & AI Applications

> **Business Logic:** > Anytime a total is calculated from quantities and rates, it is a dot product. For example, calculating total daily revenue at Lucky Dairy. 
> * **Sales Vector ($S$):** `[500L Milk, 45kg Butter, 20kg Cheese]`
> * **Price Vector ($P$):** `[120 Rs, 900 Rs, 1200 Rs]`
> Total Revenue = $S \cdot P = (500 \times 120) + (45 \times 900) + (20 \times 1200)$.

> **AI/ML Use Case:**
> 1. **Neural Networks (Perceptrons):** A single artificial neuron takes input data $\mathbf{x}$, multiplies it by learned weights $\mathbf{w}$, and sums it up. This is a dot product: $z = \mathbf{w} \cdot \mathbf{x} + b$.
> 2. **Cosine Similarity (Search/NLP):** To find how similar two sentences are, an AI converts them into vectors. The dot product of these vectors determines their similarity score.
> 3. **Attention Mechanisms (LLMs):** Transformers (like ChatGPT) determine context by calculating dot products between "Query" and "Key" vectors to see how much "attention" words should pay to one another.
