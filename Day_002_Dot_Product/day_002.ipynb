# Day 002: The Dot Product (Scalar Product)


## 1. Objective
> The **Dot Product** is the engine of similarity and reduction in linear algebra. Unlike element-wise operations (Addition, Hadamard) that return a new vector, the dot product takes two vectors of equal length collapses them into a **single scalar number**, representing *how much those two vectors align or interact with each other*.

## 2. Mathematical Foundation
> The dot product computes the sum of the products of the corresponding entries of two sequences of numbers.

Let $A$ and $B$ be two vectors of length $n$.

**The Algebraic Definition (Computation):**
$$A \cdot B = \sum_{i=1}^{n} A_i B_i = (A_1 \times B_1) + (A_2 \times B_2) + \dots + (A_n \times B_n)$$

**The Geometric Definition (Intuition):**
It measures the directional agreement between two vectors in space, where $\theta$ is the angle between them:
$$A \cdot B = \|A\| \|B\| \cos(\theta)$$
* If $\theta = 0^\circ$, $\cos(0) = 1$ (Maximum alignment).
* If $\theta = 90^\circ$, $\cos(90) = 0$ (Orthogonal / no similarity).

## 3. Real-World & AI Applications

> **Business Logic:** > Anytime a total is calculated from quantities and rates, it is a dot product. For example, calculating total daily revenue at Lucky Dairy. 
> * **Sales Vector ($S$):** `[500L Milk, 45kg Butter, 20kg Cheese]`
> * **Price Vector ($P$):** `[120 Rs, 900 Rs, 1200 Rs]`
> Total Revenue = $S \cdot P = (500 \times 120) + (45 \times 900) + (20 \times 1200)$.

> **AI/ML Use Case:**
> 1. **Neural Networks (Perceptrons):** A single artificial neuron takes input data $\mathbf{x}$, multiplies it by learned weights $\mathbf{w}$, and sums it up. This is a dot product: $z = \mathbf{w} \cdot \mathbf{x} + b$.
> 2. **Cosine Similarity (Search/NLP):** To find how similar two sentences are, an AI converts them into vectors. The dot product of these vectors determines their similarity score.
> 3. **Attention Mechanisms (LLMs):** Transformers (like ChatGPT) determine context by calculating dot products between "Query" and "Key" vectors to see how much "attention" words should pay to one another.

## 4. Algorithmic Strategy
> * Verify that both input vectors have the exact same length (raise a `ValueError` if they do not).
> * Pair up the corresponding elements from Vector A and Vector B.
> * Multiply each pair together.
> * Sum all of those individual products into a single running total.
> * **Pythonic shortcut:** Use `zip()` to pair elements, a generator expression for the multiplication, and the built-in `sum()` function to accumulate the total.

## 5. Implementation

```python
import numpy as np


class Vector:
    def __init__(self, elements):
        """Initialize with a list of numbers."""
        self.elements = elements
        
    def __repr__(self):
        return f"Vector({self.elements})"
    
    def __add__(self, other):
        """Element-wise addition."""
        if len(self.elements) != len(other.elements):
            raise ValueError("Mathematical Error: Vectors must be the same length.")
        
        summed = [a+b for a,b in zip(self.elements, other.elements)]
        return Vector(summed)
    
    def __mul__(self, scalar):
        """Element-wise scalar multiplication."""
        scaled = [a*scalar for a in self.elements]
        return Vector(scaled)
    
    
    def hadamard(self, other):
        """Element-wise multiplication."""
        if len(self.elements) != len(other.elements):
            raise ValueError("Mathematical Error: Vectors must be the same length.")
        hadamard_elements = [a*b for a,b in zip(self.elements, other.elements)]
        return Vector(hadamard_elements)
    
    # Day-2 implementation
    def dot(self, other):
        """
        Calculate the dot product of two vectors.
        Returns a Scalar (single number), not a VECTOR object.
        """
        if len(self.elements) != len(other.elements):
            raise ValueError("Mathematical Error: Vectors must be the same length.")
        
        # Multiply corresponding elements and sum them all up.
        return sum(a*b for a,b in zip(self.elements, other.elements))
```

## 6. Verification

```python
# --- 1. Define the Data ---
# Let's use the Kathmandu sales and pricing example
# Inventory: [Milk, Butter, Cheese, Yogurt]
ktm_sales_list = [500, 150, 45, 80]
prices_list = [120, 1000, 1000, 150]

# Instantiate custom vectors
ktm_sales = Vector(ktm_sales_list)
prices = Vector(prices_list)

# Instantiate NumPy arrays for verification
np_ktm_sales = np.array(ktm_sales_list)
np_prices = np.array(prices_list)

# --- 2. Calculate ---
# Scratch Calculation (our class):
revenue_scratch = ktm_sales.dot(prices)

# NumPy calculation:
revenue_np = np.dot(np_ktm_sales, np_prices)

# --- 3. Verify & Output ---
print(f"Total Revenue (Scratch): {revenue_scratch}")
print(f"Total Revenue (NumPy): {revenue_np}")


# Assert strict equality (since we are using whole numbers, == is safe):
assert revenue_scratch == revenue_np , "Dot product logic failed!"
print("✅ Dot Product verification passed! Custom logic matches NumPy perfectly.")
```

## 7. Complexity Analysis
- **Time Complexity:** `O(n)` *where $n$ is the number of elements in the vector. We must iterate through every single element pair exactly once to compute their product*
- **Space Complexity:** `O(1)` *constant space. Unlike vector addition or the Hadamard product, which require allocating memory for a completely new vector of size $n$, the dot product simply accumulates its results into a single floating-point number (scalar) in memory*
