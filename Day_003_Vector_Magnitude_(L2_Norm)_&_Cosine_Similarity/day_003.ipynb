# Day 003: Vector Magnitude (L2 Norm) & Cosine Similarity


## 1. Objective
> To calculate **geometric length of a vector (Magnitude or L2 Norm)** and use it alongside the dot product to measure the exact **directional angle (Cosine Similarity)** between two vectors, revealing *how similar they are regardless of their size*. 

## 2. Mathematical Foundation
> **i. Vector Magnitude (The L2 Norm):**  
> The physical length of a vector in multidimensional space. It is calculated using the generalized *Pythagorean Theorem*.
> Let $A$ be a vector of length $n$:
> $$\|A\| = \sqrt{A_1^2 + A_2^2 + \dots + A_n^2} = \sqrt{\sum_{i=1}^{n} A_i^2}$$

> **ii. Cosine Similarity:**  
> The measure of similarity between two vectors, calculated by the cosine of the angle $\theta$ between them. It ranges from `-1` (exact opposites) to `1` (exact match), with `0` meaning entirely orthogonal (unrelated).
> $$\text{Cosine Similarity} = \cos(\theta) = \frac{A \cdot B}{\|A\| \|B\|}$$



## 3. Real-World & AI Applications
> **Business Logic:** > Cosine Similarity identifies the *behavioral patterns* regardless of volume. Suppose you want to categorize wholesale clients for Lucky Dairy based on their ratio of Milk to Butter to Cheese purchases.
> * **Client A (Small Cafe):** Buys `[10L, 2kg, 1kg]`
> * **Client B (Large Hotel):** Buys `[100L, 20kg, 10kg]`  
> Notice that:  
> \[B = 10A\]  
> Every purchase quantity of Client B is exactly 10 times that of Client A.  
> 
> If you just calculate raw distance **(Euclidean Distance)**, these clients look vastly different. But with *Cosine Similarity*, the angle between them is $0^\circ$ (a score of `1.0`). Their buying patterns are identical, meaning the same product recommendations will work for both.  
> Both customers have the same purchasing behavior. So if customers with this pattern often also buy:    
> * $\text{Chocolate syrup}$  
> * $\text{Whipped cream}$  
> * $\text{Premium coffee filters}$  
> then those recommendations can be made to both customers.
>
> Euclidean distance measures the **absolute difference** between two vectors.  
> Formula: $$ d(A, B) = \sqrt{\sum_{i=1}^{n} (A_i - B_i)^2} $$


> **AI/ML Use Cases:**  
> i. **Recommendation Engines**: Netflix turns you into a vector (based on what you watch) and movies into vectors. It recommends the movies that have the *high cosine similarity* to your user vector.  
> ii. **Natural Language Processing (NLP)**: Words and entire documents are embedded as massive vectors (eg: Word2Vec, OpenAI embeddings). To check if a search query matches a document, the AI compute their Cosine Similarity.  
> iii. **Computer Vision**: Comparing the extracted feature vector of a scanned face against a database of known faces to authorize a login.

## 4. Algorithmic Strategy
> **For Magnitude:**  
> * Iterate through every element in the vector.  
> * Square each element and sum them up into a single total.  
> * Return the square root of that total (using Python's `math.sqrt`).  
>
> **For Cosine Similarity:**  
> * Calculate the dot product of Vector A and Vector B (using our Day 2 method).  
> * Calculate the magnitude of Vector A.  
> * Calculate the magnitude of Vector B.  
> * Multiply the two magnitude together.  
> * Divide the dot product by the multiplied magnitudes.  
> **Edge Case:** Catch division by zero if either vector is completely empty (magnitude of 0) and raise an error.

## 5. Implementation

```python
import numpy as np
import math
class Vector:
    def __init__(self, elements):
        """Initialize with the list of elements.
        """
        self.elements = elements 
    
    def __repr__(self):
        return f"Vector({self.elements})"
    
    def dot(self, other):
        """Calculates the dot product of two vectors (Day 2)."""
        if len(self.elements) != len(other.elements):
            raise ValueError("Mathematical Error: Vectors must be the same length.")
        
        return sum(a*b for a,b in zip(self.elements, other.elements))
    
    
    # Day-3 Implementation-----
    def magnitude(self):
        """
        Calculate the L2 NOrm (Euclidean length) of the vector.
        Returns a float.
        """
        sum_of_squares = sum(a**2 for a in self.elements)
        return math.sqrt(sum_of_squares)
    
    def cosine_similarity(self, other):
        """
        Calculates the cosine of the angle between two vectors.
        Returns a float between -1.0 and 1.0.
        """
        if len(self.elements) != len(other.elements):
            raise ValueError("Mathematical Error: Vectors must be the same length to compare similarity.")
        
        mag_self = self.magnitude()
        mag_other = other.magnitude()
        
        if mag_self == 0 or mag_other == 0:
            raise ValueError("Mathematical Error: Cannot compute similarity for a ZERO-Vector.")

        dot_product = self.dot(other)
        
        return dot_product/(mag_self*mag_other)
```

## 6. Verification

```python
# --- 1. Define the Data ---
# Let's test the Lucky Dairy wholesale client logic.
# [Milk (L), Butter (Kg), Cheese (Kg)]

cafe_purchases = [10, 2, 1]  # small volume
hotel_purchases = [100, 20, 10]  # Huge volume, identical ratio
bakery_purchases = [5, 50, 5]  # Totally different ratio


v_cafe = Vector(cafe_purchases)
v_hotel = Vector(hotel_purchases)
v_bakery = Vector(bakery_purchases)


np_cafe = np.array(cafe_purchases)
np_hotel = np.array(hotel_purchases)
np_bakery = np.array(bakery_purchases)


# --- 2. Calculate using our class ---
mag_cafe_scratch = v_cafe.magnitude()
sim_cafe_hotel = v_cafe.cosine_similarity(v_hotel)
sim_cafe_bakery = v_cafe.cosine_similarity(v_bakery)

# --- 3. Calculate using NumPy ---
mag_cafe_np = np.linalg.norm(np_cafe)

# NumPy doesn't have a direct 1D cosine similarity function, so we build the formula.
np_sim_cafe_hotel = np.dot(np_cafe, np_hotel)/(np.linalg.norm(np_cafe) * np.linalg.norm(np_hotel))
np_sim_cafe_bakery = np.dot(np_cafe, np_bakery)/(np.linalg.norm(np_cafe) * np.linalg.norm(np_bakery))

# --- 4. Verify & Output ---
print("---- using custom class ----")
print(f"Magnitude of Cafe Vector: {mag_cafe_scratch:.4f}")
print(f"Similarity (Cafe vs Hotel):  {sim_cafe_hotel:.4f} (1.0 means identical pattern)")
print(f"Similarity (Cafe vs Bakery): {sim_cafe_bakery:.4f}\n")

print("---- using NumPy ----")
print(f"Magnitude of Cafe Vector: {mag_cafe_np:.4f}")
print(f"Similarity (Cafe vs Hotel):  {np_sim_cafe_hotel:.4f} (1.0 means identical pattern)")
print(f"Similarity (Cafe vs Bakery): {np_sim_cafe_bakery:.4f}\n")

# Assert using math.isclose due to floating-point precision variations
assert math.isclose(mag_cafe_scratch, mag_cafe_np, rel_tol=1e-9), "Magnitude logic failed!"
assert math.isclose(sim_cafe_hotel, np_sim_cafe_hotel, rel_tol=1e-9), "Cosine Similarity logic failed!"

print("✅ Magnitude and Cosine Similarity verification passed!")
```

## 7. Complexity Analysis
- **Time Complexity:** `O(n)`. Calculating the magnitude requires one pass through the array. Calculating Cosine Similarity requires three separate $O(n)$ operations (one for the dot product, two for the magnitudes). Since $O(3n)$ simplifies algebraically to $O(n)$, the algorithm scales linearly with the size of the vectors. 
- **Space Complexity:** `O(1)` constant space. We are not creating any new arrays. We are simply down into single `float` values (the magnitude and the similarity score) in memory.
