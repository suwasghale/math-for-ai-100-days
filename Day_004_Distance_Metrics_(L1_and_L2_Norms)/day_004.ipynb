## 2. Mathematical Foundation
> Distance metrics quantify the geometric separation between two coordinate vectors $\mathbf{a}$ and $\mathbf{b}$ in an $n$-dimensional vector space $\mathbb{R}^n$.  
>
> **i. Manhattan Distance ($L_1$ Norm):**  
> Measures the distance traveling only along axis-parallel segments. It is the sum of absolute differences.  
> $$d_{L1}(\mathbf{a}, \mathbf{b}) = \|\mathbf{a} - \mathbf{b}\|_1 = \sum_{i=1}^{n} |a_i - b_i|$$
> 
> **ii. Euclidean Distance ($L_2$ Norm):**  
> Measures the straight-line distance between two points. It squares individual coordinate differences, heavily penalizing single large deviations.  
> $$d_{L2}(\mathbf{a}, \mathbf{b}) = \|\mathbf{a} - \mathbf{b}\|_2 = \sqrt{\sum_{i=1}^{n} (a_i - b_i)^2}$$
>
> **iii. Chebyshev Distance ($L_\infty$ Norm):**  
> Also known as the maximum metric. It measures distance along the single coordinate axis where the absolute difference is the greatest.  
> $$d_{L\infty}(\mathbf{a}, \mathbf{b}) = \|\mathbf{a} - \mathbf{b}\|_\infty = \max_{i} (|a_i - b_i|)$$


## 3. Real-World & AI Applications
> **Business Logic (Logistics & Urban Routing):** In a dense urban grid, vehicles cannot pass through buildings; they must follow streets. Calculating physical distance or ETA across gridded roads relies heavily on the **Manhattan ($L_1$) distance** rather than straight-line distance.
>
> **AI/ML Use Case (Robust Feature Selection & Computer Vision):**  
> >**$L_1$ vs $L_2$ Regularization:** In regression models, $L_1$ regularization (Lasso) drives uninformative feature weights strictly to ZERO, performing automatic feature selection, whereas $L_2$ regularization (Ridge) shrinks weights uniformly.  
> **Image Processing:** $L_\infty$ distance is utilized in adversarial machine learning to measure the maximum imperceptible pixel modification an attacker can make to trick an image classification network.
