# Day 004: Distance Metrics (L1 and L2 Norms)


## 1. Objective
> *To implement and analyze vector norms ($L_1$, $L_2$, and $L_\infty$) and understand how different geometric constraints alter distance calculations in multi-dimensional space.*

## 2. Mathematical Foundation
> Distance metrics quantify the geometric separation between two coordinate vectors $\mathbf{a}$ and $\mathbf{b}$ in an $n$-dimensional vector space $\mathbb{R}^n$.  
>
> **i. Manhattan Distance ($L_1$ Norm):**  
> Measures the distance traveling only along axis-parallel segments. It is the sum of absolute differences.  
> $$d_{L1}(\mathbf{a}, \mathbf{b}) = \|\mathbf{a} - \mathbf{b}\|_1 = \sum_{i=1}^{n} |a_i - b_i|$$
> 
> **ii. Euclidean Distance ($L_2$ Norm):**  
> Measures the straight-line distance between two points. It squares individual coordinate differences, heavily penalizing single large deviations.  
> $$d_{L2}(\mathbf{a}, \mathbf{b}) = \|\mathbf{a} - \mathbf{b}\|_2 = \sqrt{\sum_{i=1}^{n} (a_i - b_i)^2}$$
>
> **iii. Chebyshev Distance ($L_\infty$ Norm):**  
> Also known as the maximum metric. It measures distance along the single coordinate axis where the absolute difference is the greatest.  
> $$d_{L\infty}(\mathbf{a}, \mathbf{b}) = \|\mathbf{a} - \mathbf{b}\|_\infty = \max_{i} (|a_i - b_i|)$$


## 3. Real-World & AI Applications
> **Business Logic (Logistics & Urban Routing):** In a dense urban grid, vehicles cannot pass through buildings; they must follow streets. Calculating physical distance or ETA across gridded roads relies heavily on the **Manhattan ($L_1$) distance** rather than straight-line distance.
>
> **AI/ML Use Case (Robust Feature Selection & Computer Vision):**  
> >**$L_1$ vs $L_2$ Regularization:** In regression models, $L_1$ regularization (Lasso) drives uninformative feature weights strictly to ZERO, performing automatic feature selection, whereas $L_2$ regularization (Ridge) shrinks weights uniformly.  
> **Image Processing:** $L_\infty$ distance is utilized in adversarial machine learning to measure the maximum imperceptible pixel modification an attacker can make to trick an image classification network.

## 4. Algorithmic Strategy
> * Ensure input vectors are processed as dense NumPy arrays of identical dimension.  
> * Utilize element-wise vector operations to subtract $\mathbf{b}$ from $\mathbf{a}$ simultaneously across all dimensions.  
> * Apply matching element-wise transformations: `np.abs()` for $L_1$, squaring (`**2`) and `np.sqrt()` for $L_2$, and `np.max()` combined with `np.abs()` for $L_\infty$.

## 5. Implementation

```python
import numpy as np

def l1_distance (a: np.ndarray, b: np.ndarray)->float:
    """ 
    Calculates the Manhattan Distance (L1 Norm) between two vectors.
    """
    if a.shape != b.shape: 
        raise ValueError("Input vectors must have identical dimensions.")
    return float(np.sum(np.abs(a - b)))

def l2_distance(a: np.ndarry, b: np.ndarray) -> float:
    """ 
    Calculates the Euclidean Diatance (L2 Norm) between two vectors.
    """
    if a.shape != b.shape: 
        raise ValueError("Input vectors must have identical dimensions.")
    return float(np.sqrt(np.sum((a - b)**2)))


def l_infinity_distance(a: np.ndarray, b: np.ndarray) -> float: 
    """
    Calculates the Chebyshev Distance (L-infinity Norm) between two vectors.
    """
    if a.shape != b.shape: 
        raise ValueError("Input vectors must have identical dimensions.")
    return float(np.max(np.abs(a - b)))
```

## 6. Verification

```python
# --- Rigorous Validation Matrix ---

# Test Case: Arbitrary 4D Vector Space
vector_a = np.array([2.5, -1.0, 4.0, 0.0])
vector_b = np.array([0.5, 3.0, 1.5, -2.0])

# Compute custom distances
custom_l1 = l1_distance(vector_a, vector_b)
custom_l2 = l2_distance(vector_a, vector_b)
custom_linf = l_infinity_distance(vector_a, vector_b)

# Validate directly against Numpy C-backend references.
numpy_l1 = np.linalg.norm(vector_a - vector_b, ord=1)
numpy_l2 = np.linalg.norm(vector_a - vector_b, ord=2)
numpy_linf = np.linalg.norm(vector_a - vector_b, ord=np.inf)

print("--- Distance Verification Engine ---")
print(f"L1 (Manhattan)   -> Custom: {custom_l1:<8.4f} | NumPy: {numpy_l1:<8.4f}")
print(f"L2 (Euclidean)   -> Custom: {custom_l2:<8.4f} | NumPy: {numpy_l2:<8.4f}")
print(f"L-Inf (Chebyshev)-> Custom: {custom_linf:<8.4f} | NumPy: {numpy_linf:<8.4f}")

# Assert exact numerical parity
assert np.isclose(custom_l1, numpy_l1), "L1 validation mismatch!"
assert np.isclose(custom_l2, numpy_l2), "L2 validation mismatch!"
assert np.isclose(custom_linf, numpy_linf), "L-Infinity validation mismatch!"

print("\n✅ Geometric calculations verified perfectly across all norm orders.")
```

## 7. Complexity Analysis
- **Time Complexity:** $\mathcal{O}(n)$
  *Why:* Every metric requires a linear pass over the $n$ items in the array to compute element-wise differences ($a_i - b_i$) before performing scalar reduction operations (summation or maximum).
- **Space Complexity:** $\mathcal{O}(n)$
  *Why:* Allocation occurs during vector subtraction, producing an internal temporary delta array of size $n$ before final aggregation.
