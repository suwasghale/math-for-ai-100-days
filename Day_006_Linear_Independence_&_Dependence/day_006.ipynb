## 2. Mathematical Foundation
> **A. The Formal Definition:**  
> A set of vectors $\{\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_k\}$ is linearly independent if the only way to combine them to get the zero vector is by multiplying them all by zero.  
>   
> Mathematically, the equation:  
> $$ c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \dots + c_k\mathbf{v}_k = \mathbf{0} $$
> has **only the trivial solution** ($c_1 = 0, c_2 = 0, \dots, c_k = 0$)  
>
> **B. The Intuitive Definition (The "Span" Test):**  
> * **Linearly Dependent:** At least one vector in your set is just a linear combination of the others. It lies inside the existing Span. It adds zero new dimensions or geometric information.  
> * **Linearly Independent:** Every single vector adds a brand new dimension to the Span. None of them can be built using the others.  
>
> **C. Geometric Intuition:**  
> * **2 Vectors:** They are dependent if they lie on the exact same line (they are parallel). They are independent if they form a 2D plane.  
> * **3 Vectors:** They are dependent if they all lie flat on the same 2D plane. They are independent if they expand outward to fill a 3D volume.
