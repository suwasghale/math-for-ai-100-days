> In code, a Tensor is implemented as an N-dimensional array. NumPy gives us crucial properties to understand our mathematical structures without printing out millions of numbers:

> * `.shape`: *Returns a tuple representing the exact size of each dimension* `(Rows, Columns, Channels)`.  
> * `.size`: *Returns the total number of scalar elements inside the entire tensor*.
