## 2. Mathematical Foundation
> Matrix-vector multiplication projects a vector from one dimensional space into another using a weight matrix. It computes the dot product between the rows of the matrix and the column of the vector. 
> * **The Shape Rule:** To multiply a matrix $W$ by a vector $x$, the number of columns in $W$ must exactly match the number of rows in $x$. If $W \in \mathbb{R}^{m \times n}$ and $x \in \mathbb{R}^{n \times 1}$, the resulting vector $y$ will have shape $(m \times 1)$.  
> **The Core Operation:** Each element $y_i$ of the output vector is the sum of the products of the $i$-th row of $W$ and the vector $x$:  
> $$y_i = \sum_{j=1}^{n} W_{i,j} x_j$$
> * **The Linear Equation ($Wx + b$):** After multiplication, an intercept term (bias vector $b$ of shape $m \times 1$) is added using element-wise addition (broadcasting) to shift the output.

## 3. Real-World & AI Applications  
> * **Business Logic (Inventory Valuation):** If a local dairy business holds stock for 4 different items (e.g., milk, yogurt, paneer, butter) across 3 regional distribution hubs, $W$ is a $3 \times 4$ matrix. If the current market price of those 4 products is represented by a $4 \times 1$ vector $x$, computing $Wx$ instantly calculates the total monetary value held at each of the 3 hubs.  
> 
> * **AI/ML Use Case (The Neural Network Forward Pass):** In deep learning, an artificial neuron doesn't just add numbers—it weighs them. $x$ is the input data, $W$ represents the learned importance of each input, and $b$ is the activation threshold. Computing $y = Wx + b$ is the exact mathematical operation inside every Dense/Linear layer in computer vision or language models.

## 4. Algorithmic Strategy
> * **Dimension Alignment:** Always verify `W.shape[1] == x.shape[0]`. In NumPy, a 1D array `(n,)` behaves flexibly and acts as a column vector `(n, 1)` during multiplication.  
> * **The `@` Operator:** Use Python's native matrix multiplication operator `@` (or `np.dot()`) to compute the dot product. Do not use `*`, which performs element-wise scalar multiplication.  
> * **Broadcasting the Bias:** After computing `W @ x`, simply add the bias vector `b`. NumPy will automatically ensure the shapes align without requiring nested loops.

## 7. Complexity Analysis
> - **Time Complexity:** $\mathcal{O}(m \cdot n)$   
> **Why:** To compute the dot product of an $m \times n$ matrix with an $n \times 1$ vector, the algorithm must perform $n$ multiplications and additions for every single one of the $m$ rows. Under the hood, highly optimized C-libraries parallelize this computation.  
>  
>  - **Space Complexity:** $\mathcal{O}(m)$   
> **Why:** The resulting output vector $y$ requires a new array of size $m$ to store the computed activation values. The input vector $x$ is not duplicated during the `@` operation.
