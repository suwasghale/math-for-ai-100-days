## 2. Mathematical Foundation
> **A. The Matrix Transpose ($\mathbf{A}^T$)**  
> The transpose reflects a matrix across its main diagonal, swapping its rows and columns.  
> - If a matrix $\mathbf{A}$ has a shape of $(m \times n)$, its transpose $\mathbf{A}^T$ will strictly have a shape of $(n \times m)$.  
> - Element-wise definition:$$(\mathbf{A}^T)_{i,j} = \mathbf{A}_{j,i}$$
> **Core Properties of Transposition:**  
> - **The Double Flip:** Forcing a second transpose returns the original matrix:$$(\mathbf{A}^T)^T = \mathbf{A}$$
> - **The "Socks and Shoes" Rule:** The transpose of a matrix product reverses the order of multiplication:$$(\mathbf{A}\mathbf{B})^T = \mathbf{B}^T\mathbf{A}^T$$
> **B. The Identity Matrix ($\mathbf{I}$)** An Identity Matrix is a square matrix ($n \times n$) where all elements on the main diagonal are 1, and all other elements are 0.  
> - **Multiplicative Neutrality:** Multiplying any valid matrix by an Identity matrix preserves the original data perfectly:$$\mathbf{A}\mathbf{I} = \mathbf{A} \quad \text{and} \quad \mathbf{I}\mathbf{A} = \mathbf{A}$$(Note: $\mathbf{A}$ must be a square matrix for both equations to hold with the exact same $\mathbf{I}$, otherwise the dimensions of $\mathbf{I}$ must adjust to match $\mathbf{A}$'s inner dimensions).
