# Day 013: Matrix Inverse & Singular Matrices


## 1. Objective
> To compute matrix inverses for reversing linear transformations, solve systems of linear equations using $\mathbf{x} = \mathbf{A}^{-1}\mathbf{b}$, and programmatically catch singular matrices where information loss prevents mathematical reversal.

## 2. Mathematical Foundation
> The inverse of a square matrix $\mathbf{A}$, denoted as $\mathbf{A}^{-1}$, reverses the transformation applied by $\mathbf{A}$.  
> $$\mathbf{A}\mathbf{A}^{-1} = \mathbf{I} \quad \text{and} \quad \mathbf{A}^{-1}\mathbf{A} = \mathbf{I}$$
>- **The Determinant ($\det(\mathbf{A})$):** A scalar factor that represents the volume scaling of the matrix transformation. If $\det(\mathbf{A}) = 0$, the volume has been flattened (e.g., 3D squished to 2D).   
>- **Singularity:** A matrix with a determinant of $0$ is called a singular matrix. It has no inverse because reversing a "*flattened*" space is mathematically impossible (division by zero).

## 3. Real-World & AI Applications
>- **Business Logic (Reconstructing Hidden Variables):** Suppose you are reviewing yesterday's local production logs. You know the exact daily revenue (Rs. 11,600) and the total liters of milk consumed in processing (135 Liters). However, the notebook tracking the actual units of Paneer and Butter produced is missing.You know the constants: Paneer sells for Rs. 800 and takes 5L of milk per unit. Butter sells for Rs. 1200 and takes 20L of milk per unit.This is a linear system: $\mathbf{A}\mathbf{x} = \mathbf{b}$. By computing the inverse of the constant matrix ($\mathbf{A}^{-1}$) and multiplying it by your outputs ($\mathbf{b}$), you instantly reverse-engineer exactly how many units of each product were made.
>
>- **AI/ML Use Case (Normal Equations in Linear Regression):** When fitting a line of best fit through data using the exact algebraic method rather than Gradient Descent, the formula to find the optimal weights ($\theta$) is $\theta = (\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T\mathbf{y}$. The matrix inverse is required to solve for the perfect parameters in one shot.

## 4. Algorithmic Strategy
>- **NumPy Linear Algebra Module:** Never write an $N \times N$ inverse algorithm from scratch. Use `np.linalg.inv(A)`  
>- **Catching Errors:** Always be prepared for `np.linalg.LinAlgError`. If your data contains redundant features (e.g., Feature B is just Feature A converted to different units), your matrix will be singular and the code will crash.  
>- **Numerical Stability:** In deep learning production systems, calculating inverses for very large matrices is computationally expensive and prone to floating-point rounding errors. We often use approximate iterative solvers instead of direct inverses.

## 5. Implementation

```python
import numpy as np

# Use Case: Reversing the Dairy Production Data
# Row 1: Selling Price [Paneer: 800, Butter: 1200]
# Row 2: Milk Required [Paneer: 5L,  Butter: 20L]
A = np.array([[800.0, 1200.0],
              [  5.0,   20.0]])

# Known Outputs [Total Revenue, Total Milk Used]
b = np.array([11600.0, 135.0])

print("--- 1. Reversing the System (x = A_inv @ b) ---")
# 1. Check determinant
det_A = np.linalg.det(A)
print(f"Determinant of A: {det_A:.2f}")

if det_A !=0:
    # 2. compute inverse
    A_inv = np.linalg.inv(A)
    print("\nInverse Matrix (A_inv):")
    print(A_inv)
    
    # 3. Solve for hidden inputs (x)
    x = A_inv @ b
    print(f"\nReconstructed Data: {x[0]:.0f} units of Paneer, {x[1]:.0f} units of Butter")
    
    # 4. Prove A @ A_inv = I
    identity_check = A @ A_inv 
    print("\nProof of Neutrality (A @ A_inv):")
    print(np.round(identity_check, 2))  # Rounding to clean up floating point artifacts
```

## 6. Verification

```python
print("\n--- 2. Handling Singular Matrices ---")
# Redundant data: Row 2 is exactly Row 1 multiplied by 1.5
B = np.array([[ 2.0,  4.0],
              [ 3.0,  6.0]])

print("Matrix B:")
print(B)
print(f"Determinant of B: {np.linalg.det(B):.2f}")

try:
    B_inv = np.linalg.inv(B)
except np.linalg.LinAlgError as e:
    print(f"\nCRITICAL ERROR CAUGHT: {e}")
    print("Why? The data lacks independent variance. The transformation flattened the space.")
```

### Visualization: Rewinding the vector

```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# Data Setup
# ==========================================

# Original Input Vector
x = np.array([1.0, 1.0])

# Transformation Matrix
A = np.array([[ 2.0,  1.0],
              [-1.0,  1.5]])

# Compute Forward Pass
transformed_x = A @ x

# Compute Inverse and Reverse Pass
A_inv = np.linalg.inv(A)
reverted_x = A_inv @ transformed_x

# ==========================================
# Plotting
# ==========================================

fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title("Geometric View: Forward Transformation and Inverse Reversal", fontsize=14, pad=15)
ax.axhline(0, color='grey', linewidth=0.5, zorder=0)
ax.axvline(0, color='grey', linewidth=0.5, zorder=0)
ax.grid(True, linestyle='--', alpha=0.6)

origin = np.array([0, 0])

# 1. Plot Original Vector
ax.quiver(*origin, *x, angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.01, label='Original Input (x)')

# 2. Plot Transformed Vector
ax.quiver(*origin, *transformed_x, angles='xy', scale_units='xy', scale=1, 
           color='orange', width=0.01, label='Transformed (A @ x)')

# 3. Plot Reverted Vector (drawn slightly thinner to show it perfectly overlaps the original)
ax.quiver(*origin, *reverted_x, angles='xy', scale_units='xy', scale=1, 
           color='green', width=0.005, label='Reverted (A_inv @ Transformed)')

# Add explanatory text
ax.text(x[0] + 0.1, x[1], 'Start/End', color='blue', fontsize=12, weight='bold')
ax.text(transformed_x[0] + 0.1, transformed_x[1], 'Midpoint', color='orange', fontsize=12, weight='bold')

ax.set_xlim(-1, 4)
ax.set_ylim(-1, 4)
ax.set_aspect('equal')
ax.legend(loc='upper left')

plt.show()
```

## 7. Complexity Analysis
- **Time Complexity:** $\mathcal{O}(n^3)$ for calculating the inverse of an $N \times N$ matrix using standard decomposition methods. This is why AI avoids full matrix inversion for models with millions of parameters
- **Space Complexity:** $\mathcal{O}(n^2)$ to store the new inverted matrix in memory.
