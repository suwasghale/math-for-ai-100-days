## 2. Mathematical Foundation
> The inverse of a square matrix $\mathbf{A}$, denoted as $\mathbf{A}^{-1}$, reverses the transformation applied by $\mathbf{A}$.  
> $$\mathbf{A}\mathbf{A}^{-1} = \mathbf{I} \quad \text{and} \quad \mathbf{A}^{-1}\mathbf{A} = \mathbf{I}$$
>- **The Determinant ($\det(\mathbf{A})$):** A scalar factor that represents the volume scaling of the matrix transformation. If $\det(\mathbf{A}) = 0$, the volume has been flattened (e.g., 3D squished to 2D).   
>- **Singularity:** A matrix with a determinant of $0$ is called a singular matrix. It has no inverse because reversing a "*flattened*" space is mathematically impossible (division by zero).
