# Day 014: Solving Systems of Linear Equations (Finding the unknown algebraically)


## 1. Objective
> To frame real-world multi-variable problems as matrix equations ($\mathbf{A}\mathbf{x} = \mathbf{b}$) and solve them programmatically to find unknown quantities or optimal intersections.

## 2. Mathematical Foundation
> A system of linear equations can be elegantly compressed into matrix notation.  
> $$\mathbf{A}\mathbf{x} = \mathbf{b}$$
>
> To solve for the unknown variables vector $\mathbf{x}$, we multiply both sides by the inverse of the coefficient matrix $\mathbf{A}$.  
> $$\mathbf{x} = \mathbf{A}^{-1}\mathbf{b}$$
> * **Geometric Meaning:** Solving the system finds the exact point in space where multiple lines (2D), planes (3D), or hyperplanes (N-D) intersect.  
> * **Singularity Condition:** If $\det(\mathbf{A}) = 0$, the inverse does not exist. Geometrically, this means the lines are strictly parallel (no solution) or perfectly overlapping (infinite solutions).

## 3. Real-World & AI Applications
>* **Business Logic (Resource Constraints):** Consider optimizing inventory at Lucky Dairy. You produce Yogurt and Butter. You know exactly how much total milk you have today (1000L) and your total daily labor budget (Rs. 5000). 
  If Yogurt takes 2L of milk and Rs. 5 of labor per batch, and Butter takes 10L of milk and Rs. 15 of labor per batch, how many exact batches of each can you make to perfectly max out your resources? This translates perfectly into a 2-equation linear system that outputs the exact production targets.
>
>* **AI/ML Use Case:** In electrical engineering and circuit design (often simulated via code before building hardware), Kirchhoff's laws generate massive systems of linear equations to find the current running through every wire. In machine learning, solving $\mathbf{A}\mathbf{x} = \mathbf{b}$ is the foundational step in exact Linear Regression (the Normal Equation).


## 4. Algorithmic Strategy
> * **The Golden Rule of Code:** **Never explicitly invert a matrix if you just need to solve a system.** While mathematically $\mathbf{x} = \mathbf{A}^{-1}\mathbf{b}$ is correct, calculating $\mathbf{A}^{-1}$ on a computer is slow and highly prone to floating-point rounding errors.  
> * **The Solution:** Use NumPy's specialized solver: `np.linalg.solve(A, b)`. It uses efficient factorization (like LU decomposition) under the hood to find $\mathbf{x}$ without ever calculating the inverse.

## 5. Implementation

```python
import numpy as np

# ==========================================
# 5. Implementation & Real-World Use Case
# ==========================================

# Business Scenario: Maxing out Daily Resources
# Equation 1 (Milk Constraint): 2*(Yogurt) + 10*(Butter) = 1000 Liters
# Equation 2 (Labor Constraint): 5*(Yogurt) + 15*(Butter) = 5000 Rupees

# Coefficient Matrix (A): [Milk reqs] \n [Labor reqs]
A = np.array([[2.0, 10.0 ],
              [5.0, 15.0]])

# Constant Vector (b): [Total Milk, Total Labor ]
b = np.array([1000.0, 5000.0])

print("--- 1. Solving the production system ---")
# Using the optimized linear solver
x = np.linalg.solve(A, b)

print(f"optimzed linear solver: {x}")

print(f"Optimal Production Target: ")
print(f"Yogurt Batches: {x[0]:.0f}")
print(f"Butter Batches: {x[1]:.0f}\n")

# Verify the math
total_milk_used = (2 * x[0]) + (10 * x[1])
total_labor_used = (5 * x[0]) + (15 * x[1])
print(f"Verification -> Milk used: {total_milk_used}L | Labor used: Rs. {total_labor_used}")
```

## 6. Verification

```python
import time
# solver vs explicit invers

print("\n--- 2. Performance: Solver vs Inverse ---")
# Create a massive 1000x1000 system of equations
A_large = np.random.rand(1000, 1000)
b_large = np.random.rand(1000)

# Method 1: The explicit (and dangerous) inverse method
start_inv = time.time()
x_inv = np.linalg.inv(A_large) @ b_large
time_inv = time.time() - start_inv
print(f"Time using np.linalg.inv:   {time_inv:.4f} seconds")

# Method 2: The optimized solver method
start_solve = time.time()
x_solve = np.linalg.solve(A_large, b_large)
time_solve = time.time() - start_solve
print(f"Time using np.linalg.solve: {time_solve:.4f} seconds")

# Why we use solve(): It is generally faster, and significantly more numerically stable!
```

### Visualization: The intersection of realities

```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# Data Setup
# ==========================================

# We have two equations for (x = Yogurt, y = Butter):
# 1: 2x + 10y = 1000  =>  y = (1000 - 2x) / 10
# 2: 5x + 15y = 5000  =>  y = (5000 - 5x) / 15

# Generate x values (Yogurt batches)
x_vals = np.linspace(0, 1000, 400)

# Calculate corresponding y values (Butter batches) for each line
y_milk = (1000 - 2 * x_vals) / 10
y_labor = (5000 - 5 * x_vals) / 15

# The solved intersection point (from our python solver logic)
intersect_x = 175
intersect_y = 65

# ==========================================
# Plotting
# ==========================================

fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title("Geometric Solution: Intersection of Constraints", fontsize=15, pad=15)

# Plot the lines
ax.plot(x_vals, y_milk, label='Milk Constraint Line (2x + 10y = 1000)', color='blue', linewidth=2)
ax.plot(x_vals, y_labor, label='Labor Constraint Line (5x + 15y = 5000)', color='orange', linewidth=2)

# Plot the exact solution intersection
ax.plot(intersect_x, intersect_y, 'ro', markersize=8, label=f'Solution Intersection ({intersect_x}, {intersect_y})')

# Add visual guide lines to the axes
ax.vlines(intersect_x, 0, intersect_y, color='red', linestyle='--', alpha=0.5)
ax.hlines(intersect_y, 0, intersect_x, color='red', linestyle='--', alpha=0.5)

# Formatting
ax.set_xlim(0, 600)
ax.set_ylim(0, 350)
ax.set_xlabel('Yogurt Batches (x)', fontsize=12)
ax.set_ylabel('Butter Batches (y)', fontsize=12)
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.grid(True, linestyle=':', alpha=0.7)
ax.legend(loc='upper right')

plt.show()
```

## 7. Complexity Analysis
- **Time Complexity:** $\mathcal{O}(n^3)$ for direct solvers (LU decomposition / Gaussian elimination). For massive systems (e.g., millions of variables), AI uses iterative approximation methods (like Gradient Descent) which are much faster than $\mathcal{O}(n^3)$.
- **Space Complexity:**  $\mathcal{O}(n^2)$ to store the coefficient matrix.
